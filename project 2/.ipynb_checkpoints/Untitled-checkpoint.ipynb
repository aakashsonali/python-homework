```python
#install the dependiencies
import yfinance as yf
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

```python
rbc = yf.Ticker("ry")

```

```python
data = yf.download("ry", start="2019-02-28", end="2020-02-28")
```

```python
data
```

```python
#visulaize the close price data
plt.figure(figsize=(16,8))
plt.title('RBC')
plt.xlabel('Days')
plt.ylabel('Close Price Cad ($)')
plt.plot(data['Close'])
plt.show()
```

```python
#create a dataframe
df= data[['Close']]
df.head()
```

```python
#Create a variable to predict 'x ' days out in the future
future_days = 25
#Create a new column (target) 
df['Prediction'] = df[['Close']].shift(-future_days)
df.tail()
```

```python
#create a feature data set (x ) and convert it to a munpy array and remove the last 'x'  rows/days
X = np.array(df.drop(['Prediction'], 1))[: -future_days]
print(X)
```

```python
# Create a target data set (y) and convert it to numpy array and get all of target values except the last 
y = np.array(df['Prediction'])[: -future_days]
print(y)
```

```python
#Split the data into traing and testing
x_train, x_test, y_train, y_test = train_test_split (X , y, test_size = 0.25)
```

```python
# Create the models
# Create the Decision tree regressor model
tree= DecisionTreeRegressor().fit(x_train, y_train)
# Create the linear regression model
lr= LinearRegression().fit(x_train, y_train)
```

```python
# Create the last 'x' rows of future data set
x_future= df.drop(['Prediction'], 1) [: -future_days]
x_future= x_future.tail(future_days)
x_future= np.array(x_future)
x_future
```

```python
# shoe the model tree prediction
tree_prediction= tree.predict(x_future)
print(tree_prediction)
# show the model linear regression prediction
lr_prediction = lr.predict(x_future)
print(lr_prediction)
```

```python
#Visiualize the data

predictions = tree_prediction

valid = df[X.shape[0]: ]
valid['Predictions']= predictions
plt.figure (figsize= (16,8))
plt.title('Model')
plt.xlabel('Days')
plt.ylabel('close price cad ($) ')
plt.plot(df['Close'])
plt.plot(valid[['Close' ,'Predictions']])
plt.legend(['orig', 'Val', 'Pred'])
plt.show()
```

```python
predictions = lr_prediction

valid = df[X.shape[0]: ]
valid['Predictions']= predictions
plt.figure (figsize= (16,8))
plt.title('Model')
plt.xlabel('Days')
plt.ylabel('close price cad ($) ')
plt.plot(df['Close'])
plt.plot(valid[['Close' ,'Predictions']])
plt.legend(['orig', 'Val', 'Pred'])
plt.show()
```
